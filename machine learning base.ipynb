```python
# # importing pandas and the main taxonomic data 
import pandas as pd
rawData = pd.read_csv(r'E:\the main dirctory\Machine Learning project\train_data.csv')
```

```python
# # removing some of the columns
rawData = rawData.drop(columns=['blood pressure','smoking','diabetic foot','obesity','sex'])
```

```python
# # creating output set


#outputData = rawData['disas stat']
outputData = rawData['diabet']
```

```python
# # removing the colum "disease state" from inputData



inputData = rawData.drop(columns = ['FBS'])
```

```python
# # start machine learning


from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(inputData, outputData)
```

```python
for i in range (1,10):
# # EVALUATING  spliting data



    from sklearn.model_selection import train_test_split
    from sklearn.metrics import accuracy_score

    inputData_train,inputData_test, outputData_train, outputData_test =  train_test_split(inputData, outputData, test_size = 0.2)

    model.fit(inputData_train, outputData_train)
    predictions = model.predict(inputData_test)

    score = accuracy_score (outputData_test, predictions)
    print(score)
```

```python
# save the trained model



import sklearn as external
import joblib

joblib.dump(model, 'the first version of my trained model_FBS VS diabet type.joblib')
```
